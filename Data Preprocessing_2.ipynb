```python
import pandas as pd
import re

# Load the dataset
file_path = 'C:/Users/Aman/Downloads/final sms dataset.csv'  # Replace with the actual file path
data = pd.read_csv(file_path)

# Function to extract account number from the Text column
def extract_account_number(text):
    match = re.search(r'\b(?:a/c|account|AC|a/c no.)\s*(\w+)', str(text), re.IGNORECASE)
    return match.group(1) if match else None

# Apply the function to extract account numbers
data['Account Number'] = data['Text'].apply(extract_account_number)

# Save or display the updated dataset
data.to_csv('C:/Users/Aman/Downloads/updated final sms dataset.csv', index=False)  # Save to a new file
data.head()  # Display the first few rows
```
